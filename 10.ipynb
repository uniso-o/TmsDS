```python
class HomeScenario:
    def __init__(self, motion: bool, door: bool, room: str, dict_rooms: dict):
        self.__motion = motion
        self.__door = door
        self.__room = room
        self.__dict_rooms = dict_rooms

    @property
    def motion(self):
        return self.__motion

    @property
    def door(self):
        return self.__door
        
    @property
    def room(self):
        return self.__room
        
    @property
    def dict_rooms(self):
        return self.__dict_rooms
        
    def on_motion_detected(self):
        keys = self.dict_rooms.keys()
        if self.room in keys and self.motion == True:
            print(f"Движение в {self.room}. Запись начала.")
            
    def on_door_opened(self):
        keys = self.dict_rooms.keys()
        if self.door == True and self.room in keys:
            print(f"Дверь {self.dict_rooms.get(self.room)} открыта. Оповещение отправлено.")

class AwayScenario(HomeScenario):
    def __init__(self, motion, door, room, dict_rooms):
        super().__init__(motion, door, room, dict_rooms)
        
    def on_motion_detected(self):
        keys = self.dict_rooms.keys()
        if self.room in keys and self.motion == True:
            print(f"Включена сирена, производится запись видео")
            
    def on_door_opened(self):
        keys = self.dict_rooms.keys()
        if self.door == True and self.room in keys:
            print(f"Дверь {self.dict_rooms.get(self.room)} заблокирована. Производится звонок в полицию.")

class SleepScenario(HomeScenario):
    def __init__(self, motion, door, room, dict_rooms):
        super().__init__(motion, door, room, dict_rooms)
        self.__light = 0

    def on_motion_detected(self):
        keys = self.dict_rooms.keys()
        if self.room == 'Коридор' and self.motion == True:
            print(f"Включен свет на {self.__light + 10}%")
            
    def on_door_opened(self):
        keys = self.dict_rooms.keys()
        if self.room == 'Спальня' and self.room in keys:
            print(f"Дверь {self.dict_rooms.get(self.room)} открыта. Уведомления отключены. Входная дверь заблокирована")

class PartyScenario(HomeScenario):
    def __init__(self, motion, door, room, dict_rooms):
        super().__init__(motion, door, room, dict_rooms)
        self.__sound = 100
        
    def on_motion_detected(self):
        keys = self.dict_rooms.keys()
        if (self.room == 'Кухня' or self.room == 'Спальня') and self.motion == True and self.room in keys:
            print(f"Звук уменьшен на {self.__sound - 50}%")
            
    def on_door_opened(self):
        keys = self.dict_rooms.keys()
        if self.room == 'Вход' and self.room in keys:
            print(f"Дверь {self.dict_rooms.get(self.room)} открыта. Голосовой помощник приветствует гостя")
```

```python
home = HomeScenario(True, True, "Кухня", {"Кухня": "1", "Вход": "2","Туалет": "3", "Коридор": "4", "Спальня": "5"})
home.on_motion_detected()
home.on_door_opened()
home1 = AwayScenario(True, True, "Кухня", {"Кухня": "1", "Вход": "2","Туалет": "3", "Коридор": "4", "Спальня": "5"})
home1.on_motion_detected()
home1.on_door_opened()
home2 = SleepScenario(True, True, "Коридор", {"Кухня": "1", "Вход": "2","Туалет": "3", "Коридор": "4", "Спальня": "5"})
home2.on_motion_detected()
home2.on_door_opened()
home3 = PartyScenario(True, True, "Вход", {"Кухня": "1", "Вход": "2","Туалет": "3", "Коридор": "4", "Спальня": "5"})
home3.on_motion_detected()
home3.on_door_opened()
```
