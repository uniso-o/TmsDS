```python
#матрицы
import random
import numpy as np
class Matrix_1:
    def __init__(self, rows: int ):
        self.__rows = rows
        self.__matrix = []
        #for i in range(rows):#custom
            #self.__matrix.append(int(input()))   
        for i in range(rows):#custom
            self.__matrix.append(random.randint(0, 100))
            
    def sum_elements(self):
        for i in range(len(self.__matrix)):
            result = sum(self.__matrix)
        print(result)
        return result

    @property
    def matrix(self):
        return self.__matrix
        
    def sumobject(self, matrix):
        result = []
        for x, y in zip(self.__matrix, matrix.__matrix):
            result.append(x + y)
        print(result)
        
    def multobject(self, matrix):
        result = []
        for x, y in zip(self.__matrix, matrix.__matrix):
            result.append(x * y)
        print(result)
                          
    def show_matrix(self):
        print(self.__matrix)

class Matrix_2(Matrix_1):
    def __init__(self, rows = 0, cols = 0):
        super().__init__(rows)
        self.__matrix = []
        self.__rows = rows
        self.__cols = cols 
        
    def create_matrix(self):
        for i in range(self.__rows):#order
            temp = []
            for j in range(self.__cols):
                temp.append(i * self.__rows + j + 1)
            self.__matrix.append(temp)
        print(np.array(self.__matrix))

    def sum_elements(self):
        result = 0
        for i in range(self.__rows):
            for j in range(self.__cols):
                result += self.__matrix[i][j]
        print(result)
        
    @property
    def matrix(self):
        return self.__matrix
            
    @property
    def size(self):
        return self.__rows, self.__cols 
        
    def sumobject(self, matrix):
        result = []
        if self.size == matrix.size:
            for i in range(self.__rows):
                temp = []
                for j in range(self.__cols):
                    temp.append(self.__matrix[i][j] + matrix.__matrix[i][j])
                result.append(temp)
            print(np.array(result))
        else:
            print('Размерность матриц не совпадвет!')

    def multobject(self, matrix):
        result = []
        if self.size == matrix.size:
            for i in range(self.__rows):
                temp = []
                for j in range(self.__cols):
                    temp.append(self.__matrix[i][j] * matrix.__matrix[i][j])
                result.append(temp)
            print(np.array(result))
        else:
            print('Размерность матриц не совпадвет!')

    def saddle_point(self):
        find = False
        for i in range(self.__rows):
            min_el = min(self.__matrix[i])
            j = self.__matrix[i].index(min_el)
            column = [self.__matrix[m][j] for m in range(self.__rows)]
            max_el = max(column)
            if min_el == max_el:
                print(f'Седловая точка: {min_el} в позиции ({i}, {j})')
                find = True
        if not find:
            print('Седловой точки нет!')

class Matrix_3(Matrix_2):
    def __init__(self, rows = 0, cols = 0, height = 0):
        super().__init__(rows, cols)
        self.__rows = rows
        self.__cols = cols 
        self.__height = height
        self.__matrix = []  
        
    def create_matrix(self):
        for i in range(self.__rows):#random  
            t = []
            for j in range(self.__cols):
                temp = []
                for k in range(self.__height):
                    temp.append(random.randint(0, 100))
                t.append(temp)
            self.__matrix.append(t)
        print(np.array(self.__matrix))    
        
    def sum_elements(self):
        result = 0
        for i in range(self.__rows):
            for j in range(self.__cols):
                for k in range(self.__height):
                    result += self.__matrix[i][j][k]
        print(result)   

    @property
    def matrix(self):
        return self.__matrix
            
    @property
    def size(self):
        return self.__rows, self.__cols 
        
    def sumobject(self, matrix):
        if self.size == matrix.size:
            sum = []
            for i in range(self.__rows):
                result = []
                for j in range(self.__cols):
                    temp = []
                    for k in range(self.__height):
                        temp.append(self.__matrix[i][j][k] + matrix.__matrix[i][j][k])
                    result.append(temp)
                sum.append(result)
            print(np.array(sum))  
        else:
            print('Размерность матриц не совпадвет!')

    def multobject(self, matrix):
        if self.size == matrix.size:
            mul = []
            for i in range(self.__rows):
                result = []
                for j in range(self.__cols):
                    temp = []
                    for k in range(self.__height):
                        temp.append(self.__matrix[i][j][k] * matrix.__matrix[i][j][k])
                    result.append(temp)
                mul.append(result)
            print(np.array(mul))  
        else:
            print('Размерность матриц не совпадвет!')

    def saddle_point(self):
        find = False
        for k in range(len(self.__matrix)):
            lim = self.__matrix[k] 
            for i in range(len(lim)):
                min_el = min(lim[i])
                j = lim[i].index(min_el)
                column = [lim[m][j] for m in range(len(lim))]
                max_el = max(column)
                if min_el == max_el:
                    print(f'Седловая точка: {min_el} в позиции ({i}, {j})')
                    find = True
        if not find:
            print('Седловой точки нет!')
```

```python
matrix1 = Matrix_1(4)
matrix2 = Matrix_1(4)
matrix1.show_matrix()
matrix2.show_matrix()
matrix3 = Matrix_2(3, 3)
matrix33 = Matrix_2(3, 3)
matrix3.create_matrix()
matrix33.create_matrix()
matrix4 = Matrix_3(3, 3, 3)
matrix5 = Matrix_3(3, 3, 3)
matrix4.create_matrix()
matrix5.create_matrix()
matrix1.sum_elements()
matrix3.sum_elements()
matrix4.sum_elements()
matrix2.sumobject(matrix1)
matrix2.multobject(matrix1)
matrix3.sumobject(matrix33)
matrix3.multobject(matrix33)
matrix4.sumobject(matrix5)
matrix4.multobject(matrix5)
matrix3.saddle_point()
matrix4.saddle_point()
```
