```python
import numpy as np
arr_A1 = np.array([[1, 2, 4],
                   [7, 5, -2],
                   [-2, 1, 3]])

arr_A2 = np.array([[0, 3, 4],
                   [-7, 1, -2],
                   [2, -1, 3]])

arr_A3 = np.array([[5, 2, 9],
                   [7, 0, -2],
                   [-2, 1, 6]])

arr_A4 = np.array([[1, -2, 8],
                   [7, 5, -2],
                   [-2, 3, 0]])

arr_A5 = np.array([[9, 2, 4],
                   [7, 5, -5],
                   [6, -2, 3]])

arr_B1 = np.array([[-1, 0, 3],
                   [2, -6, 1],
                   [-7, 1, 5]])

arr_B2 = np.array([[1, 2, 4],
                   [7, 0, -2],
                   [-3, -2, 5]])

arr_B3 = np.array([[1, -2, 1],
                   [7, 5, -2],
                   [-2, 8, 3]])

arr_B4 = np.array([[0, 2, 4],
                   [1, -3, -2],
                   [-2, 1, 3]])

arr_B5 = np.array([[8, 2, 4],
                   [0, 1, -3],
                   [-2, 1, 6]])

arr_C1 = np.array([[-2, 3],
                   [4, 0],
                   [7, -1]])

arr_C2 = np.array([[-3, 2],
                   [0, 4],
                   [-1, 7]])

arr_C3 = np.array([[-2, 3],
                   [4, 5],
                   [0, -1]])

arr_C4 = np.array([[0, 3],
                   [5, -4],
                   [7, -1]])

arr_C5 = np.array([[-2, 0],
                   [4, 8],
                   [1, -1]])

def summatrix(a, b):
    return np.add(a, b)
    
def submatrix(a, b):
    return np.subtract(a, b)
    
def dotmatrix(a, b):
    if len(b) == len(a[0]):
        return a.dot(b)
        
def multmatrix(a, b, c):
    if len(b) == len(a[0]):
        result = a.dot(b)
        if len(c) == len(result[0]):
            return result.dot(c)
```

```python
print(summatrix(arr_A1, arr_B1))
print(summatrix(arr_A3, arr_B3))
print(summatrix(arr_B1, arr_A1))
print(dotmatrix(arr_A1, arr_C1))
print(multmatrix(arr_A1, arr_B1, arr_C1))
```
