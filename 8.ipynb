```python
#1
import math
class Shape:
    def __init__(self, name):
        self.__name = name
        
    def area(self):
        return 'Неизвестно'
        
    def perimeter(self):
        return 'Неизвестно'
        
    def display_info(self):
        print('Фигура:', self.get_name())
        print('Площадь:', self.area())
        print('Периметр:', self.perimeter())
        
    def get_name(self):
        return self.__name
        
class Circle(Shape):
    def __init__(self, r, name = 'circle'):
        super().__init__(name)
        self.__r = r
        if self.__r <= 0:
            raise ValueError('Число должно быть положительным')
            
    def display_info(self):
        super().display_info()
        print('Радиус:', self.__r)
        
    def area(self):
        return math.pi * self.__r ** 2  
        
    def perimeter(self):
        return 2 * math.pi * self.__r
        
class Rectangle(Shape):
    def __init__(self, width, height, name = 'rectangle'):
        super().__init__(name)
        self.__width = width
        self.__height = height
        if self.__width <= 0 or self.__height <= 0:
            raise ValueError('Числа должны быть положительными')  
            
    def display_info(self):
        super().display_info()
        print('Ширина:', self.__width)
        print('Длина:', self.__height)
        
    def area(self):
        return self.__width * self.__height 
        
    def perimeter(self):
        return 2 * (self.__width + self.__height)
        
class Triangle(Shape):
    def __init__(self, a, b, c, name = 'triangle'):
        super().__init__(name)
        self.__a = a
        self.__b = b
        self.__c = c
        if self.__a <= 0 or self.__b <= 0 or self.__c <= 0:
            raise ValueError('Числа должны быть положительными')
    
    def checktriangle(self):
        if self.__a + self.__b <= self.__c or self.__a + self.__c <= self.__b or self.__c + self.__b <= self.__a:
            raise ValueError('Треугольника не существует')
            
    def perimeter(self):
        return self.__a + self.__b + self.__c
        
    def area(self):
        semip = self.perimeter() / 2
        return math.sqrt(semip * (semip - self.__a) * (semip - self.__b) * (semip - self.__c))
        
    def display_info(self):
        super().display_info()
        print('Стороны:', self.__a, self.__b, self.__c)  
     
class Square(Rectangle):
    def __init__(self, height, name = 'square'):
        super().__init__(height, height, name)
        self.__height = height
        
    def display_info(self): 
        super().display_info()
        print('Сторона квадрата:', self.__height) 
```

```python
circle = Circle(5)
circle.area()
circle.perimeter()
circle.get_name()
rectangle = Rectangle(5, 7)
rectangle.area()
rectangle.perimeter()
rectangle.get_name()
triangle = Triangle(5, 8, 9)
square = Square(5)
square.area()
square.perimeter()
square.display_info() 
circle.display_info()
triangle.display_info()
```

```python
#2

class Employee:
    def __init__(self, name, ID, date, post):
        self._name = name
        self._ID = ID
        self._post = post
        self._date = date
        
    def salary_calculate(self): 
        self._result = self._salary - self._salary / 13
        print('Зарплата с вычетом налогов:', self._result)
        
    def vacation_calculate(self):
        if self._vac_days > 0:
            vac_salary = self._result / self._vac_days
            print('Отпускные:', vac_salary)
            
    def employee_info(self):
        print(f'Информация о сотруднике:\nname:{self._name}\nID:{self._ID}\ndate:{self._date}\npost:{self._post}')
        
    def tax_fee_calculate(self):
        if self._salary - self._salary / 13 == self._result:
            print(f'Налоговый вычет:{self._salary / 13}')
        else:
            print(f'Налоговый вычет:{self._salary / 13}\nВзнос:{self._fee}')
            
    def accounting_sick_vac_days(self):
        if self._vac_days or self._sick_days > 0:
            self._attendance = self._vac_days + self._sick_days 
            print(f'Отпуск и/или больничный: {self._attendance}')
        
class FullTimeEmployee(Employee):
    def __init__(self, name, ID, date, salary, work_days = 0, vac_days = 0, sick_days = 0, fee = 0, post = 'FullTimeEmployee'):
        super().__init__(name, ID, date, post)
        self._salary = salary
        self._work_days = work_days
        self._vac_days = vac_days
        self._sick_days = sick_days
        self._fee = fee
        if self._work_days > 200:
            self._salary *= 2
        
class PartTimeEmployee(Employee):
    def __init__(self, name, ID, date, hours, rate, post = 'PartTimeEmployee'):
        super().__init__(name, ID, date, post)
        self._hours = hours
        self._rate = rate
        
    def salary_calculate(self):
        self._salary = self._hours * self._rate
        return super().salary_calculate()
        
class CommissionEmployee(Employee):
     def __init__(self, name, ID, date, month_rate, percent_sale, sale_volume, post = 'CommissionEmployee'):
        super().__init__(name, ID, date, post)
        self._month_rate = month_rate
        self._percent_sale = percent_sale
        self._sale_volume = sale_volume
         
     def salary_calculate(self):
        self._salary = self._month_rate + self._month_rate / 100 * self._percent_sale
        if self._sale_volume > 150:
            self._month_rate + self._month_rate / 2
        return super().salary_calculate()
         
class Manager(FullTimeEmployee):
    def __init__(self, name, ID, date, salary, work_days = 0, vac_days = 0, sick_days = 0, fee = 0, bonus = 0, budget = 0, post = 'Manager'):
        super().__init__(name, ID, date, salary, work_days, vac_days, sick_days, fee, post)
        self.__bonus = bonus
        self.__budget = budget 
        
    def salary_calculate(self):
        if self.__budget > 10000:
            self._salary += self.__bonus
        return super().salary_calculate()
            
```
