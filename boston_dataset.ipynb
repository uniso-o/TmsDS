```python
import seaborn as sns
import matplotlib.pyplot as plt 
plt.figure(figsize=(12,10), dpi= 80)
sns.heatmap(df.corr(), xticklabels=df.corr().columns, yticklabels=df.corr().columns, cmap='RdYlGn', center=0, annot=True)
plt.title('medv', fontsize=22)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
sns.scatterplot(data=df, x="medv", y = "crim", ax=axes[0,0])
axes[0,0].set_title("      ")
sns.scatterplot(data=df, x="medv", y = "tax", ax=axes[0,1])
axes[0,1].set_title("      ")
sns.scatterplot(data=df, x="medv", y = "rm", ax=axes[1,0])
axes[1,0].set_title("   \n     ")
sns.scatterplot(data=df, x="medv", y = "age", ax=axes[1,1])
axes[1,1].set_title("       ")
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
df = df.dropna()

x = df.drop(columns = "medv") 
y = df["medv"]            
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

print(" :", y_pred[:10])   
print(" :", y_test[:10].values)  
```
