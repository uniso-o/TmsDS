```python
def prepare_dataframe(df, text_column, custom_stopwords=['br', '<', '>', '/', 'its', "'ll", "n't", '.', ',', '....', '(', ')', '..', '...']):

    stop_words = set(stopwords.words('english'))
    stop_words.update([w.lower() for w in custom_stopwords])

    def preprocess_text(text):
        if not isinstance(text, str):
            return []
        text = text.lower()
        text = re.sub(r'[^a-zA-Z\s]', '', text)
        tokens = word_tokenize(text)
        tokens = [word for word in tokens if word not in stop_words and len(word) > 1]
        return tokens

    df_processed = df.copy()
    df_processed['processed_tokens'] = df_processed[text_column].apply(preprocess_text)
    df_processed['processed_text'] = df_processed['processed_tokens'].apply(
        lambda x: ' '.join(x) if x else ''
    )

    return df_processed
```

```python
from gensim.parsing.preprocessing import preprocess_string, STOPWORDS
custom_filters = [
    lambda x: x.lower(),                    # нижний регистр
    strip_tags,                             # удаление HTML тегов
    strip_punctuation,                      # удаление пунктуации
    strip_multiple_whitespaces,             # удаление лишних пробелов
    strip_numeric,                          # удаление цифр
    lambda x: re.sub(r'[^a-zA-Z\s]', '', x), # только буквы
    remove_stopwords                        # удаление стандартных стоп-слов
]
```

```python
def filter_with_gensim(text, filters=None):
    if not isinstance(text, str):
        return []

    if filters is None:
        filters = custom_filters

    tokens = preprocess_string(text, filters=filters)
    return tokens
```

```python
def get_document_vector(tokens, model, vector_size=100):

    word_vectors = [model.wv[word] for word in tokens if word in model.wv]

    if len(word_vectors) == 0:
        return np.zeros(vector_size)

    return np.mean(word_vectors, axis=0)

X = np.array([get_document_vector(tokens, w2v_model, 100) for tokens in sentences])
y = df_result['sentiment'].values
```

```python
def spacy_filter(text):

    if not isinstance(text, str):
        return []

    doc = spacy_model(text.lower())

    filtered_tokens = []

    for token in doc:
        if (not token.is_stop and
            not token.is_punct and
            not token.is_space and
            not token.is_digit and
            len(token.text) > 2):

            filtered_tokens.append(token.lemma_)

    return filtered_tokens
```

```python
def cleantext_filter(text):
    if not isinstance(text, str):
        return []

    cleaned = clean(
        text,
        fix_unicode=True,
        to_ascii=True,              # конвертация в ASCII
        lower=True,                 # нижний регистр
        no_numbers=True,
        no_punct=True,
        replace_with_number=""
    )

    tokens = cleaned.split()

    return tokens
```

```python
from lemminflect import getLemma
```

```python
def filter_with_lemminflect(text):
    if not isinstance(text, str):
        return []

    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()

    lemmatized = []
    for token in tokens:
        if len(token) > 2:
            lemma = getLemma(token, 'VERB')
            if not lemma:
                lemma = getLemma(token, 'NOUN')
            lemmatized.append(lemma[0] if lemma else token)
        else:
            lemmatized.append(token)

    stop_words = {'the', 'a', 'an', 'and', 'or', 'but', 'is', 'are', 'was', 'were',
                  'to', 'of', 'for', 'in', 'on', 'at', 'with', 'by', 'br', '<', '>',
                  '/', 'Its', "'ll",  "n't", '.', ',', '....', '(', ')', '..', '...'}
    lemmatized = [t for t in lemmatized if t not in stop_words and len(t) > 2]

    return lemmatized
```

```python
from textblob import TextBlob
```

```python
def textblob_filter(text):
    if not isinstance(text, str):
        return []

    blob = TextBlob(text.lower())

    stopwords = {'the', 'a', 'an', 'and', 'or', 'but', 'is', 'are', 'was', 'were',
                 'to', 'of', 'for', 'in', 'on', 'at', 'with', 'by', 'this', 'that',
                 'br', '<', '>', '/', 'Its', "'ll",  "n't", '.', ',', '....', '(', ')', '..', '...'}

    tokens = []
    for word in blob.words:
        if word not in stopwords and len(word) > 2 and word.isalpha():
            tokens.append(word.lemmatize())

    return tokens
```

```python
df_result['textblob_tokens'] = df_result['review'].apply(textblob_filter)
```

```python
from textblob import TextBlob

def textblob_sentiment_predict(text):
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity
    return 1 if polarity > 0 else 0

y_textblob = df_result['review'].apply(textblob_sentiment_predict)
y_true = df_result['sentiment'].values

accuracy_blob = accuracy_score(y_true, y_textblob)
print(f"TextBlob accuracy: {accuracy_blob:.4f}")
```
